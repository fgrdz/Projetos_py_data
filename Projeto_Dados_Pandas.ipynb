```python
import pandas as pd

arquivo = "/content/Amazon Books Data.csv"

dataframe = pd.read_csv(arquivo, nrows=50)

```

```python
dataframe= dataframe[["title", "description", "author", "rating", "price"]]
dataframe
```

```python
import numpy as np
rating = dataframe['rating']
np.median(rating) # media
```

```python
authors = dataframe['author']
```

```python
import matplotlib.pyplot as plt

plt.barh(dataframe.head()['title'], dataframe.head()['rating'])
plt.xlabel('title')
plt.ylabel('rating')
plt.show()
```

```python
plt.barh(dataframe.head()['title'],dataframe.head()['price'])
plt.xlabel('title')
plt.ylabel('price')
plt.show()
```

```python
checar_palavra = dataframe['title'].str.contains('Python')

for i, founded in enumerate( checar_palavra):
  if founded:
    print(f"Linha {i} {founded} para python no título")
# imprime todas as linhas em que os livros contém python no título
```

```python
lista_indices=[]
for i, founded in enumerate(checar_palavra):
  if founded:
    lista_indices.append(i)
lista_indices
#cria uma lista com os indices dessas linhas
```

```python
livros_python = dataframe.loc[lista_indices]['title'].tolist()
print(livros_python)
# armazena apenas os livros que contém python no título
```

```python
rating_python = dataframe.loc[lista_indices]['rating'].tolist()
list(zip(livros_python, rating_python))
# mescla as duas listas e exibe os livros e a classificação
```

```python
plt.barh(livros_python, rating_python)
plt.xlabel('title')
plt.ylabel('price')
plt.show()
```
